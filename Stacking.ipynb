```python
X = data.drop(['Survived'],axis=1)
y = data['Survived']
```

```python
from sklearn.model_selection import train_test_split
train_x , test_x,train_y,test_y = train_test_split(X,y,random_state=0)
train_x.shape , test_x.shape,train_y.shape,test_y.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
def model_prediction(model,train_x,train_y,test_x):
  model.fit(train_x,train_y)
  pred_train = model.predict(train_x)
  pred_test = model.predict(test_x)
  return pred_train,pred_test
```

```python
DT = DecisionTreeClassifier(random_state=0)
M1_train,M1_test=model_prediction(DT,train_x,train_y,test_x)
```

```python
LR = LogisticRegression(random_state=0)
M2_train,M2_test=model_prediction(LR,train_x,train_y,test_x)
```

```python
knn = KNeighborsClassifier()
M3_train,M3_test=model_prediction(knn,train_x,train_y,test_x)
```

```python
train_pred = {
    'DT': M1_train,
    'LR': M2_train,
    'knn':M3_train
}
train_pred = pd.DataFrame(train_pred)
train_pred
```

```python
test_pred = {
    'DT': M1_test,
    'LR': M2_test,
    'knn':M3_test
}
test_pred = pd.DataFrame(test_pred)
test_pred
```

```python
model= LogisticRegression(random_state=0)
model.fit(train_pred,train_y)
```

```python
model.score(test_pred,test_y)
```
