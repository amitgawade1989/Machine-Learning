```python
import pandas as pd 
import numpy as np
data = pd.read_csv('/content/drive/MyDrive/Datasets/data_cleaned.csv')
```

```python
data.head()
```

```python
X = data.drop(['Survived'],axis=1)
y = data['Survived']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(X,y,random_state=0)
```

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier()
abc.fit(train_x,train_y)
```

```python
pred_y = abc.predict(test_x)
pred_y
```

```python
abc.score(train_x,train_y)
```

```python
abc.score(test_x,test_y)
```

## hyper parameter tuning

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = AdaBoostClassifier(random_state=100,base_estimator=RandomForestClassifier(random_state=10),n_estimators=100,learning_rate=0.005)
```

```python
clf.fit(train_x,train_y)
```

```python
clf.score(train_x,train_y)
```

```python
clf.score(test_x,test_y)
```
